## Типы признаков

Признаки или фичи (features) - численное описание объектов выборки,
которое помогает модели наиболее эффективно приближать целевую зависимость. 

Типы:
    
   * численный
   
   (Пример: непрерывные значения ПЭС)
   
   
   * категориальные
       * упорядоченные (ordianl)
       
        (Пример: уровень геомагнитной активности: низкий, средний, высокий)
       
       * не упорядоченные (номинальные, nominal)
       
        (Пример: характер геомагнитной активности)       
       

То, в каком виде следует представить данные сильно зависит от моделей, которые вы собираетесь использовать.

Например, для большинства нейросетевых моделей, линейной регрессии, KNN и прочих необходимо нормировать данные.

Есть тренд, явная нестационарность. То есть мат. ожидание и дисперсия зависят от времени.
Покажем это на расчётах.

```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
    
test_stationarity(data["#Passengers"])
```

Что можно сделать?

Есть трюк. Можно вычесть тренд и штрафовать высокие значения.
Или сделать, как предложили на семинаре - разделить на тренд.

```python
moving_avg_ = pd.rolling_mean(ts,12)
ts_div = (ts - moving_avg_)/ moving_avg_
```

```python
ts_new = ts_div
```

```python
ts_new.plot()
```

```python
ts_log = np.log(ts)
moving_avg_log = pd.rolling_mean(ts_log,12)
ts_new = ts_log - moving_avg_log
plt.plot(ts_new)
```

```python
test_stationarity(ts_new.dropna())
```

```python
ts_new.dropna().hist()
```

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf

lag_acf = acf(ts_new.dropna(), nlags=20)

#Plot ACF: 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_new)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_new)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
```

```python
data['target'] = ts_new
data['month_num'] = data.index.month
data['lag6'] = data['target'].shift(6)
data['lag12'] = data['target'].shift(12)
data.dropna(inplace=True)
data.head(5)
```

```python
plt.scatter(data['lag12'], data['target'])
```

```python
train = data.query("Month < '1959-10-01'")
test = data.query("Month >= '1959-10-01'")
feature_list = ['lag12','lag6'] # 'month_num',
train_X = train[feature_list]
train_Y = train['target']
test_X = test[feature_list]
test_Y = test['target']
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(train_X, train_Y)

# Make predictions using the testing set
y_pred = regr.predict(test_X)
```

```python
regr.coef_
```

```python
plt.plot(y_pred)
```

```python
plt.plot(test_Y)
```

```python
# The mean squared error
print("Mean squared error: %.5f"
      % mean_squared_error(test_Y, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(test_Y, y_pred))
```

### Задача регрессии (предсказание диабета)

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes = datasets.load_diabetes()
X = pd.DataFrame(diabetes.data)
y = pd.Series(diabetes.target)
```

```python
print(diabetes.DESCR)
X.columns=['Age','Sex','BMI','ABP','1','2','3','4','5','6']
```

```python
# Use only one feature
diabetes_X = diabetes.data[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue', linewidth=3)
```
